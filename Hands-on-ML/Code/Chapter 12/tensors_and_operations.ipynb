tf.tensor has dtype and shape

```
t = tf.constant([[1., 2., 3.], [4., 5., 6.]])
print(t.shape)
print(t.dtype)
```

Indexing works like numpy

```
t[:, 1:]
```

all sorts of tensor operations are available

```
t + 10
```

```
tf.square(t)
```

```
t @ tf.transpose(t)
```
