```
import tensorflow as tf
import numpy as np
```

```
def cube(x):
  return x ** 3
```

```
cube(2)
```

```
cube(tf.constant(2.0))
```

```
tf_cube = tf.function(cube)
tf_cube
```

```
tf_cube(2)
```

```
tf_cube(tf.constant(2.0))
```

```
@tf.function
def tf_cube(x):
  return x ** 3
tf_cube.python_function(2)
```
